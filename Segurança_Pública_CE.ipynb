```python
import pandas as pd
import datetime
from datetime import time
import plotly as plt
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
```

```python
df_furto_2021 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Furto_12_2021.xls")
df_furto_2022 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Furto_12_2022.xls")
df_furto_2023 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Furto_10_2023.xls")

df_cvli_2021 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/CVLI_12_2021.xls")
df_cvli_2022 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/CVLI__12_2022.xls")
df_cvli_2023 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/CVLI_10_2023.xls")

df_droga_2021 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Entorpecente_12_2021.xls")
df_droga_2022 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Entorpecente_12_2022.xls")
df_droga_2023 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Entorpecente_10_2023.xls")

df_armas_2021 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Arma-de-Fogo_12_2021.xls")
df_armas_2022 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Arma-de-Fogo_2022.xls")
df_armas_2023 = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Segurança Pública/Arma-de-Fogo_10_2023.xls")


#Juntando os Dados
df_furto = pd.concat([df_furto_2021, df_furto_2022, df_furto_2023]).reset_index(drop=True)
df_cvli = pd.concat([df_cvli_2021, df_cvli_2022, df_cvli_2023]).reset_index(drop=True)
df_droga = pd.concat([df_droga_2021, df_droga_2022, df_droga_2023]).reset_index(drop=True)
df_ARMAS = pd.concat([df_armas_2021, df_armas_2022, df_armas_2023]).reset_index(drop=True)
```

# Armas de Fogo

```python
df_ARMAS['Crime'] = 'Armas'
df_ARMAS['Mes'] = df_ARMAS['Data'].dt.month
df_ARMAS['Dia'] = df_ARMAS['Data'].dt.day
```

```python
# regulando a coluna de horas
df_ARMAS['Hora'] = df_ARMAS['Hora'].apply(lambda x: str(x)[:2])
```

```python
df_ARMAS
```

```python
# Criando um DataFrame para Areas da Capital Fortaleza
df_ARMAS_C = df_ARMAS[df_ARMAS['AIS'].isin(['AIS 1','AIS 2','AIS 3','AIS 4','AIS 5','AIS 6','AIS 7','AIS 8','AIS 9', 'AIS 10',])]

# Criando um DataFrame para Areas do Interior
df_ARMAS_I = df_ARMAS[df_ARMAS['AIS'].isin(['AIS 11', 'AIS 12','AIS 13','AIS 14','AIS 15','AIS 16','AIS 17','AIS 18','AIS 19', 'AIS 20','AIS 21','AIS 22',])]

```

```python
pivot_df = df_ARMAS_C.pivot_table(index='Dia da Semana', columns='AIS', values='Quantidade', aggfunc='sum')

fig = px.imshow(pivot_df,
                labels=dict(x="Área Integrada de Segurança", y="Dia da Semana", color="Valor"),
                x=pivot_df.columns,
                y=pivot_df.index)


fig.update_layout(
    title='Quantidade de Armas Apreendias em Dia da Semana x A Área da apreensão',
    xaxis=dict(title='Área Integrada de Segurança'),
    yaxis=dict(title='Dia da Semana')
)

fig.update_layout(coloraxis=dict(colorscale='Cividis'))


fig.show()
```

```python
pivot_df = df_ARMAS_C.pivot_table(index='AIS', columns='Hora', values='Quantidade', aggfunc='sum')

fig = px.imshow(pivot_df,
                labels=dict(x="Horas com maiores indeces", y="Região", color="Valor"),
                x=pivot_df.columns,
                y=pivot_df.index)


fig.update_layout(
    title='Locais e Horas onde há maior apreensão de armas',
    xaxis=dict(title='Hora da Apreensão'),
    yaxis=dict(title='Área Integrada de Segurança '),
    width=1200,
    height=600
)

fig.update_layout(coloraxis=dict(colorscale='Plasma'))


fig.show()
```
