## Armas de Fogo

```python
#Criando a Função para padronizar os valores AIS, alguns dados apresentam "AIS 1" outros "AIS 01"
def inserir_zero(texto):
    if len(texto) < 6:
      return texto[:-1] + '0' + texto[-1]
    else:
      return texto

# Padronizando AIS
df_armas['AIS'] = df_armas['AIS'].apply(lambda x: inserir_zero(x))
```

```python
#Criando as colunas de Crime, Dia e Mês
df_armas['Crime'] = 'Armas'
df_armas['Mes'] = df_armas['Data'].dt.month
df_armas['Dia'] = df_armas['Data'].dt.day
```

```python
# Regulando a coluna de horas
df_armas['Hora'] = df_armas['Hora'].apply(lambda x: str(x)[:2])
```

As Áreas de Integradas de Segurança (AIS) são divididas em áreas do interior e da capital. Aqui há um pequeno problema de granulometria entre os dados, a capital concentra 10 AIS's e as demais estão distribuidas entre diversas cidades do interior. Enquanto umas concentram dados de bairros, outras concetram dados de cidades inteiras. Isso é um problema pois estaremos comparando dados de bairros com dados de 2, 3 , 4 ou mais cidades concentradas em uma AIS. Devido a isso, optei por separar as ánalises da capital e do interior

```python
# Criando um DataFrame para Areas da Capital Fortaleza
df_armas_C = df_armas[df_armas['AIS'].isin(['AIS 01','AIS 02','AIS 03','AIS 04','AIS 05','AIS 06','AIS 07','AIS 08','AIS 09', 'AIS 10',])]

# Criando um DataFrame para Areas do Interior
df_armas_I = df_armas[df_armas['AIS'].isin(['AIS 11', 'AIS 12','AIS 13','AIS 14','AIS 15','AIS 16','AIS 17','AIS 18','AIS 19', 'AIS 20','AIS 21','AIS 22',])]

```

Futuramente trabalharei os dados do interior, mas por hora direcionarei esse estudo para os dados da capital.

```python
df_armas_C.head()
```

Plotarei alguns mapas de calor para explorar os dados.

```python
pivot_df = df_armas_C.pivot_table(index='AIS', columns='Mes', values='Quantidade', aggfunc='sum')

fig = px.imshow(pivot_df,
                labels=dict(x="Horas com maiores indeces", y="Região", color="Valor"),
                x=pivot_df.columns,
                y=pivot_df.index)


fig.update_layout(
    title='Mês x Região',
    xaxis=dict(title='Mês'),
    yaxis=dict(title='Área Integrada de Segurança '),
    width=1200,
    height=700
)

fig.update_layout(coloraxis=dict(colorscale='Plasma'))

fig.update_yaxes(categoryorder='total ascending', categoryarray=pivot_df.index)


fig.show()
```

As regiões formadas pela AIS 3 **(Ancuri, Barroso, Coaçu, Conjunto Palmeiras, Curió, Guajeru, Jangurussu, Lagoa Redonda, Messejana, Parque Santa Maria, Paupina, Pedras e São Bento.)** são as quue apresentam no decorrer de todos os meses do ano o maior indice de apreensões de arma. Já a AIS 7 ganha destaque apenas no primeiro mês do ano.

```python
pivot_df = df_armas_C.pivot_table(index='AIS', columns='Dia da Semana', values='Quantidade', aggfunc='sum')

fig = px.imshow(pivot_df,
                x=pivot_df.columns,
                y=pivot_df.index)

fig.update_layout(
    title='Quantidade de Armas Apreendias no decorrer da semana',
    xaxis=dict(title='Dia da Semana'),
    yaxis=dict(title='Região'),
)

fig.update_layout(coloraxis=dict(colorscale='Plasma'))

fig.update_yaxes(categoryorder='total ascending', categoryarray=pivot_df.index)

fig.show()
```

```python
pivot_df = df_armas_C.pivot_table(index='AIS', columns='Hora', values='Quantidade', aggfunc='sum')

fig = px.imshow(pivot_df,
                labels=dict(x="Horas com maiores indeces", y="Região", color="Valor"),
                x=pivot_df.columns,
                y=pivot_df.index)


fig.update_layout(
    title='Locais e Horas onde há maior apreensão de armas',
    xaxis=dict(title='Hora da Apreensão'),
    yaxis=dict(title='Área Integrada de Segurança '),
    width=1200,
    height=600
)

#Paleta de cores
fig.update_layout(coloraxis=dict(colorscale='Plasma'))

#Degrade
fig.update_yaxes(categoryorder='total ascending', categoryarray=pivot_df.index)

fig.show()
```

Todas as regiões apresentam um aumento das apreensões de aramas de fogo no período vespertino, tendo seu ápice no período noturno. Os horários entre as 02:00 da manhã e as 08:00 são os que apresentam o menor numero de apreensões.
