# Crimes

Todos os DF's  vem com seus dados com pouca alteração a serem feitas

```python
#Criando uma Função para regular os parametros de cada Datarfame
def pre_df(df,nome):
  #Criando a Função para padronizar os valores AIS, alguns dados apresentam "AIS 1" outros "AIS 01"
  def inserir_zero(texto):
    if len(texto) < 6:
      return texto[:-1] + '0' + texto[-1]
    else:
      return texto
  df['AIS'] = df['AIS'].apply(lambda x: inserir_zero(x))

  # Retirando apanas a hora da ocorrência
  df['Hora'] = df['Hora'].apply(lambda x: str(x)[:2])

  #Criando Colnas relevantes
  df['Crime'] = str(nome)
  df['Mes'] = df['Data'].dt.month
  df['Dia'] = df['Data'].dt.day
```

As Áreas de Integradas de Segurança (AIS) são divididas em áreas do interior e da capital. Aqui há um pequeno problema de granulometria entre os dados, a capital concentra 10 AIS's e as demais estão distribuidas entre diversas cidades do interior. Enquanto umas concentram dados de bairros, outras concetram dados de cidades inteiras. Isso é um problema pois estaremos comparando dados de bairros com dados de 2, 3 , 4 ou mais cidades concentradas em uma AIS. Devido a isso, optei por separar as ánalises da capital e do interior

```python
# Criando um DataFrame para Areas da Capital Fortaleza
df_armas_C = df_armas[df_armas['AIS'].isin(['AIS 01','AIS 02','AIS 03','AIS 04','AIS 05','AIS 06','AIS 07','AIS 08','AIS 09', 'AIS 10',])]

# Criando um DataFrame para Areas do Interior
df_armas_I = df_armas[df_armas['AIS'].isin(['AIS 11', 'AIS 12','AIS 13','AIS 14','AIS 15','AIS 16','AIS 17','AIS 18','AIS 19', 'AIS 20','AIS 21','AIS 22',])]

```

Dentre os crimes abordados, aqueles que geram violência física ou grave ameaça a pessoa são os que mais preocupam a sociedade, dentre eles estão os Crimes Violentos Letais e Intensionais - CVLI, Crimes Violentos Patrimoniais - CVP, Maria da Penha - M_Penha e Sexuais.  

Para criar o heatmap e dar uma visão da criminalinade fortalezense vou usar como base esses quatro crimes. E por que não usar todos? simples, algumas regiões bastantes violentas tem poucos registro de furtos quando comparada com outras, isso se deve ao controle das facções criminosas do estado nessas regiões.

```python
df1 = df[df['Crime'].isin(['CVLI', 'CVP', 'M_Penha', 'Sexuais'])]
pivot_df = df1.pivot_table(index='AIS', columns='Hora', values='Crime', aggfunc='count')

fig = px.imshow(pivot_df,
                labels=dict(x="Horas com maiores indeces", y="Região", color="Valor"),
                x=pivot_df.columns,
                y=pivot_df.index)


fig.update_layout(
    title='Locais e Horas onde há maiores indices de CVLI',
    xaxis=dict(title='Hora'),
    yaxis=dict(title='Área Integrada de Segurança '),
    width=1200,
    height=600
)

fig.update_layout(coloraxis=dict(colorscale='Plasma'))

#Degrade
#fig.update_yaxes(categoryorder='total ascending', categoryarray=pivot_df.index)


fig.show()
```
